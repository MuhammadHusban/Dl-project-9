# Creating a Cnn for Male and Femal Classification

# Importing Libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Activation,Dense,Flatten,Conv2D,MaxPool2D,BatchNormalization
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import glob
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
%matplotlib inline
import pandas as pd
```

# Setting up our Data into

# Train
# Test

```python
os.chdir("Machine-Learning\gender")
```

```python
os.chdir('../../')
```

```python
train_path = 'Machine-Learning\gender/Training'
test_path = 'Machine-Learning\gender/Validation'
```

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input) \
    .flow_from_directory(directory=train_path,target_size=(224,224),classes=['female','male'],batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input) \
    .flow_from_directory(directory=test_path,target_size=(224,224),classes=['female','male'],batch_size=10)
```

```python
assert train_batches.n == 47009
assert test_batches.n == 11649
assert train_batches.num_classes == test_batches.num_classes == 2
```

# Ploting train images

```python
imgs,labels = next(train_batches)
```

```python
def plotImages(images_arr):
    fig,axes = plt.subplots(1,10,figsize = (10,20))
    axes = axes.flatten()
    for img,ax in  zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(imgs)
print(labels)
```

# Creating Model for cllassification

```python
model = Sequential([
    Conv2D(filters=32,kernel_size=(3,3),activation='relu',padding='same',input_shape = (224,224,3)),
    MaxPool2D(pool_size=(2,2),strides=2),
    Conv2D(filters=64,kernel_size=(3,3),activation='relu',padding='same'),
    MaxPool2D(pool_size=(2,2),strides=2),
    Flatten(),
    Dense(units = 2,activation='softmax'),
])
```

```python
model.summary()
```

# Compiling model

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss = 'categorical_crossentropy',metrics=['accuracy'])
```

# Now fitting into train batch

```python
model.fit(x = train_batches,epochs = 10,verbose = 2)
```

# Plotting test batches

```python
test_imgs,test_labels = next(test_batches)
plotImages(test_imgs)
print(test_labels)
```

# Predicting 

```python
predictions = model.predict(x=test_batches,verbose = 0)
```

```python
np.round(predictions)
```
